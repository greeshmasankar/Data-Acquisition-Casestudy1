```python
Q1=df['Population 2024'].quantile(0.25)
Q1
```

```python
Q3=df['Population 2024'].quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound
```

```python
upper_bound
```

```python
Outliers = df[(df['Population 2024'] < lower_bound) | (df['Population 2024'] > upper_bound)]
Outliers
```
